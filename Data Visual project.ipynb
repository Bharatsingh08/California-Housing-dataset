```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
data = pd.read_csv("/content/sample_data/california_housing_test.csv")
data.head()
```

```python
data.plot(kind="scatter", x="longitude", y="latitude")
```

```python
sns.jointplot(x="longitude", y="latitude", data=data, size=5)
```

```python
sns.FacetGrid(data, hue="households", size=5) \
   .map(plt.scatter, "longitude", "latitude") \
   .add_legend()
```

```python
sns.boxplot(x="longitude", y="latitude", data=data)
```

```python
sns.violinplot(x="population", y="households", data=data, size=4)
```

```python
df = pd.DataFrame(data)
  
X = list(df.iloc[:,0])
Y = list(df.iloc[:,1])
plt.bar(X, Y, color='b')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
data.hist(bins=10)
plt.xlim([0,100])
plt.ylim([50,500])
plt.title("Data")
plt.show()
```
